```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
diabetes = pd.read_csv("Diabetes.csv")
print(diabetes)
```

```python
diabetes.head()
```

```python
diabetes.describe()
```

```python
diabetes.info()
```

```python
diabetes.corr()
```

```python
d=diabetes.loc[(diabetes['Glucose']!=0) & (diabetes['BloodPressure']!=0) & (diabetes['SkinThickness']!=0) & 
               (diabetes['Insulin']!=0) & (diabetes['BMI'])]
```

```python
print(d.shape)
print(diabetes.shape)
```

```python
#Dataframe 1: diabetes (768 rows) #Dataframe 2: d (392 rows)- average of Glucose, BP, ST, Insulin and BMI
```

```python
d.describe()
```

```python
d.info()
```

```python
diabetes['Glucose'].replace(0,d['Glucose'].mean(),inplace=True)
diabetes['BloodPressure'].replace(0,d['BloodPressure'].mean(),inplace=True)
diabetes['SkinThickness'].replace(0,d['SkinThickness'].mean(),inplace=True)
diabetes['Insulin'].replace(0,d['Insulin'].mean(),inplace=True)
diabetes['BMI'].replace(0,d['BMI'].mean(),inplace=True)
```

```python
diabetes.describe()
```

```python
diabetes.corr()
```

```python
x=diabetes.iloc[:,0:8]
```

```python
y=diabetes.iloc[:,-1]
```

```python
print(y)
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
print('predicted value')
print(y_pred)
print('Actual value')
print(y_test)
```

```python
d2=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
d2.head()
```

```python
y_test
```

```python
cf=confusion_matrix(y_test,y_pred)
print('Confusion Matrix')
print(cf)
```

```python
print('Classification Report for Testing Dataset')
print(classification_report(y_test,y_pred))
```

```python
y_pred_train=lr.predict(x_train)
```

```python
print('Classification Report for Training Dataset')
print(classification_report(y_train,y_pred_train))
```
